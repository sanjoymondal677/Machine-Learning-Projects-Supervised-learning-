### Feature selection
### By heatmap we will check how the numerical independents variables are  correlated with Species or if we get any independents are correlated with each other take one of theme for model.
#### Here data is small so we can take all the independent variables for the model.

```python
plt.figure(figsize=(22,8))
cor = data.corr()
sns.heatmap(cor,annot = True)
plt.show()
```

```python
def correlation(dataset,threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_fea = correlation(data,0.9)  # here if we want equal to 0.9 threshold or more we will get only 'PetalWidthCm' we can drop it but here data is small so we can implement our model without drop that column.
```

```python
# LABEL_ENCODING("Species")
# data
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
data['Species'] = labelencoder.fit_transform(data['Species'])
data
```

```python
x = data.drop('Species',axis=1) #features data
```

```python
y = data['Species'] # labeled data or out put
```

#### split the data set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

### Logistic regression (for binary classification)

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(solver='liblinear')
logmodel.fit(X_train,Y_train)
```

```python
# Let see how accurate our model for prediction
```

```python
pred = logmodel.predict(X_test)
```

```python
print(pred)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test,pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(Y_test,pred))
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy Score is :", accuracy_score(Y_test,pred))
```
